```python
def queryNames(word):
    '''
    Returns set of names likely to be relevant to the user
    '''
    adjustments = [word, word.lower(), word.lower() + 's', word.capitalize(), word.capitalize() + 's', word[:-1] + 'ing']
    names = m.CodeToNames
    matches = set()
    for Id in names:
        words = names[Id].split()
        for adjustment in adjustments:
            if adjustment in words:
                matches.add(names[Id])
    return matches

def queryIds(series_title):
    '''
    FOR USE BY SUPPLY MAVEN SYSTEM
    Returns timeseries for a given series title
    '''
    for Id in m.CodeToNames:
        if m.CodeToNames[Id] == series_title:
            return m.MasterPrices[Id]
        
        
        
```

```python
match = queryNames('merchandise')
match
```

```python
s = []
for i in match:
    sx = queryIds(i)
    s.append(sx)
s
```
